```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../csv/signes.csv', sep=',')
```

```python
data
```

```python
data = data.drop(columns="nom_video")
data
```

```python
y = data["classe"]
data = data.drop(columns="classe")
```

```python
# réorganiser les données par ordre alphanumérique
data = data.reindex(sorted(data.columns, key=lambda x: ("_".join(x.split("_")[:-2]),int(x.split("_")[-1]))), axis=1)
data
```

```python
# add the class column
data["classe"] = y
```

```python
data
```

```python
# création du dataframe final
arrays = [np.array([i for i in range(0,data.shape[0]) for j in range(0, 875)]), np.array([i for j in range(0, data.shape[0]) for i in range(0,875)])]
index = pd.MultiIndex.from_arrays(arrays, names=('instance', 'frame'))
final_data = pd.DataFrame(0, index = ["MAR", "EAR_left", "EAR_right", "EAR_mean", "HOP_gd", "HOP_hb", "EBR", "PERCLOS", "classe"], columns=index)
final_data = final_data.T
```

```python
final_data
```

```python
d = data[[f"MAR_{i}" for i in range(0,875)]]
d = d.to_numpy().flatten()
d.shape
```

```python
for sign in ["MAR", "EAR_left", "EAR_right", "EAR_mean", "HOP_gd", "HOP_hb", "EBR", "PERCLOS"]:
    data2 = data[[f"{sign}_{i}" for i in range(0,875)]]
    d = data2.to_numpy()
    d = d.flatten()
    final_data[sign] = d
    
final_data
# , "EBR", "PERCLOS"
```
