```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier CSV avec pandas
nom_fichier_csv = 'dataframe_final.csv'  # Remplacez par le nom de votre fichier CSV
dataframe = pd.read_csv(nom_fichier_csv)

# Extraire les colonnes EAR_{i} pour i de 0 à 149
colonnes_ear = [f'EAR_moyen_{i}' for i in range(150)]
donnees_ear = dataframe[colonnes_ear]

# Calculer la moyenne des valeurs EAR pour chaque colonne
ear_moyen = donnees_ear.mean()

# Tracer les valeurs moyennes EAR en fonction de i
plt.figure(figsize=(10, 6))
plt.plot(range(150), ear_moyen, marker='o', linestyle='-', color='b')

plt.xlabel('i')
plt.ylabel('Valeurs moyennes EAR')
plt.title('Graphique des valeurs moyennes EAR en fonction de i')
plt.grid(True)
plt.show()
```
